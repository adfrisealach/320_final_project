### Import 2000

```python
# import 1990-2000
pop_2000 = pd.read_csv('./dirty/state_pop/state_pop_2000-2010.csv', skiprows=3, skipfooter=10, engine='python')
```

```python
# Create new column names
column_name = ['state', '2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010']

# rename columns
pop_2000.columns = column_name

# Drop last year - Import in next df
pop_2000 = pop_2000.drop(columns='2010')
```

### Import 2010s

```python
# import 1990-2000
pop_2010 = pd.read_csv('./dirty/state_pop/state_pop_2010-2020.csv', skiprows=3, skipfooter=8, engine='python')
```

```python
pop_2010.head(5)
```

```python
pop_2010 = pop_2010.drop(columns=['Estimates Base','2010'])
```

```python
# Initial column drop

# Create new column names
column_name = ['state', '2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']

# rename columns
pop_2010.columns = column_name
```

### Merge Population ranges together

```python
pop = pop_1990.merge(pop_2000, how='inner', on='state')
```

```python
pop = pop.merge(pop_2010, how='inner', on='state')
```

```python
pop.sample(5)
```

## Restructure

```python
population = pop.melt(id_vars='state')
```

```python
population
```

```python
population = population.rename(columns={'variable':'year','value':'population'})
```

## Write out

```python
population.to_csv('./clean/population_annual_state.csv', index=False)
```
